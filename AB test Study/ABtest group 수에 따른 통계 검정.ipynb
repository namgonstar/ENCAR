# AB Test 그룹 수에 따른 통계 검정 방법론과 분석 프로세스

## ✅ 그룹 수에 따른 통계 검정 요약

| 그룹 수 | 대표 검정법 | 목적 |
|---------|-------------|------|
| 2개     | Z-test, Chi-square test, t-test, Bayesian A/B test | 두 그룹 간 차이 검정 |
| 3개 이상 | ANOVA, Chi-square test, Post-hoc test (Tukey HSD 등), Bayesian A/B/N test | 여러 그룹 간 차이 여부 및 어떤 그룹이 유의한지 확인 |

---

## 🎯 1. 그룹이 2개일 때: A vs B

### 대표 검정 방법

| 검정 방법             | 사용 조건                      | 해석                                |
|----------------------|-------------------------------|-------------------------------------|
| Z-test               | 전환율 (비율 데이터) 비교      | 두 그룹 전환율 차이의 유의성        |
| t-test (독립표본)    | 연속형 지표 평균 비교           | 평균 차이 검정                      |
| Chi-square test      | 범주형 지표 (클릭/미클릭 등)    | 독립성 검정                         |
| Bayesian A/B test    | 확률 분포 기반 해석             | `P(A > B)` 등의 확률 기반 해석       |

### 분석 프로세스

1. **가설 수립**
   - H0: A와 B 그룹의 전환율(또는 평균)에 차이가 없다.
   - H1: A와 B 그룹의 전환율(또는 평균)에 차이가 있다.

2. **통계 검정 수행**
   - Z-test, t-test, chi-square 중 적절한 방법 선택

3. **p-value 확인**
   - 일반적으로 p < 0.05일 때 통계적으로 유의함

4. **효과 크기 및 신뢰구간 확인 (옵션)**

5. **결론 도출**

## 🎯 2. 그룹이 3개 이상일 때: A vs B vs C (또는 그 이상)

### 대표 검정 방법

| 검정 방법             | 사용 조건                       | 해석                                   |
|----------------------|--------------------------------|----------------------------------------|
| One-way ANOVA        | 연속형 지표 평균 비교           | 모든 그룹 평균이 동일한지 검정         |
| Chi-square test      | 범주형/비율 비교                | 그룹 간 비율이 동일한지 검정           |
| Bayesian A/B/N test  | 확률 분포 기반 해석             | `P(group_i > group_j)` 확률 기반 해석  |
| Post-hoc test        | ANOVA 이후 그룹 간 개별 비교     | 어떤 쌍의 그룹 간 차이가 유의한지 분석 |

## 이 상황에서 범주형 지표에 대해 통계 검정 하려면 z-test 를 쓰고 다중검정 보정 하면 된다.

### 분석 프로세스

1. **가설 수립**
   - H0: 모든 그룹의 평균(또는 비율)이 같다.
   - H1: 적어도 하나의 그룹은 다르다.

2. **통계 검정 수행**
   - ANOVA 또는 chi-square test 실행

3. **p-value 확인**
   - 유의미한 차이(p < 0.05)가 있을 경우 다음 단계로 진행

4. **Post-hoc 분석**
   - Tukey HSD, Bonferroni, FDR 등 사용

5. **다중 검정 보정**
   - 다수의 비교를 할 때 Type I error를 줄이기 위해 보정 필요

6. **해석 및 결론 도출**
