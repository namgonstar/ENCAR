각 용어는 실험 설계, 분석, 해석 단계에서 매우 중요하게 쓰입니다.

## ✅ A/B 테스트 주요 통계 용어 정리

### 📘 1. **가설 (Hypothesis)**

* **Null Hypothesis (H₀)**: A와 B 간에 차이가 없다.
* **Alternative Hypothesis (H₁)**: A와 B 간에 유의미한 차이가 있다.

---

### 📘 2. **p-value**

* **정의**: 귀무가설(H₀)이 참일 때, 현재 관측된 결과보다 극단적인 결과가 나올 확률
* **해석**:

  * p < 0.05 → 귀무가설 기각 (통계적으로 유의미함)
  * p ≥ 0.05 → 귀무가설 채택 (차이가 우연일 수 있음)

---

### 📘 3. **유의수준 (Significance Level, α)**

* 일반적으로 **0.05 (5%)** 사용
* p-value가 α보다 작으면 → 귀무가설 기각

---

### 📘 4. **검정력 (Statistical Power, 1 - β)**

* 참된 대립가설(H₁)을 올바르게 발견할 확률
* 일반적으로 **0.8 이상**이 이상적
* 검정력에 영향을 주는 요소:

  * 샘플 수
  * 효과 크기 (Effect size)
  * 유의수준

---

### 📘 5. **1종 오류 (Type I Error, False Positive)**

* 실제로는 차이가 없는데, 있다고 판단하는 오류
* 확률 = **α (유의수준)**

### 📘 6. **2종 오류 (Type II Error, False Negative)**

* 실제로는 차이가 있는데, 없다고 판단하는 오류
* 확률 = **β**

---

### 📘 7. **효과 크기 (Effect Size)**

* A와 B 간의 **실질적인 차이의 크기**
* 예: 전환율 차이 2%p, 평균 구매 금액 차이 등
* 크기가 클수록 p-value가 작아질 확률이 높음

---

### 📘 8. **신뢰구간 (Confidence Interval)**

* 예: “A 그룹의 전환율은 12% ± 1% (95% 신뢰수준)”
* 실제 모수가 이 구간에 있을 확률이 95%라는 뜻

---

### 📘 9. **z-test / t-test**

* 두 집단 간 평균이나 비율 비교를 위한 통계검정

  * **z-test**: 샘플 수가 많고 정규분포 가정 가능할 때
  * **t-test**: 샘플 수가 작거나 분산 모를 때

---

### 📘 10. **Multiple Testing (다중검정 문제)**

* 여러 A/B 실험을 동시에 하면 **우연히 유의한 결과가 나올 확률이 증가**
* 해결책:

  * **Bonferroni Correction**
  * **FDR (False Discovery Rate) 조절**
