#  Z-test에서 각 그룹 간의 쌍별 비교마다 p-value가 따로 계산되는데도, 왜 다중검정 보정을 해야 하는지에 대해

아주 좋은 질문이에요. Z-test에서 **각 그룹 간의 쌍별 비교마다 p-value가 따로 계산되는데도, 왜 다중검정 보정을 해야 하는지**에 대해 헷갈릴 수 있습니다. 아래에서 개념부터 이유까지 깔끔하게 정리해드릴게요.

---

## ✅ 1. 쌍별 비교(p-value)는 독립이 아님

예를 들어, 그룹이 A, B, C, D 네 개라면, 우리는 다음 6가지 쌍을 비교합니다:

* A vs B
* A vs C
* A vs D
* B vs C
* B vs D
* C vs D

이 **6개의 비교는 모두 같은 실험 데이터셋**에서 나온 것입니다. 즉, 완전히 독립된 실험이 아니라 **같은 데이터에서 반복적으로 유의성 검정을 하고 있는 셈**이죠.

---

## ✅ 2. 왜 다중검정 문제가 발생할까?

### 🎯 핵심 문제: **우연히 유의해 보일 확률이 늘어난다**

* 유의수준(α)을 0.05로 설정하면, **실제로 차이가 없더라도 약 5% 확률로 p-value가 0.05 미만이 나올 수 있습니다.**
* 쌍별 비교가 많아질수록 이 \*\*우연히 유의하다고 나오는 오류(False Positive)\*\*가 누적됩니다.

예를 들어:

| 쌍별 비교 개수 | 잘못된 유의성 발견 확률 |
| -------- | ------------- |
| 1개 비교    | 5% (정상적 범위)   |
| 6개 비교    | 약 26%         |
| 10개 비교   | 약 40% 이상      |

즉, 여러 쌍을 비교하면서도 각각 α = 0.05로 따지면, **전체 실험에서 잘못된 결론을 낼 위험이 급격히 증가**합니다.

---

## ✅ 3. 그래서 다중검정 보정이 필요하다

### 💡 목적:

> 여러 비교를 수행했을 때도 전체적으로 **유의수준이 유지되도록** 보정하는 것.

### 🎓 대표적인 방법들:

* **Holm-Bonferroni 보정**: 보수적인 방식. family-wise error rate (FWER)을 통제.
* **Benjamini-Hochberg (FDR)**: 유연한 방식. False Discovery Rate를 통제.

---

## ✅ 4. 예시로 비교

| 비교쌍    | Uncorrected p-value | Holm 보정 후 p-value | FDR 보정 후 p-value |
| ------ | ------------------- | ----------------- | ---------------- |
| A vs B | 0.012               | 0.048             | 0.030            |
| A vs C | 0.020               | 0.060             | 0.030            |
| B vs C | 0.050               | 0.100             | 0.050            |

* **Uncorrected p-value**만 보면 다 유의해 보일 수 있음
* 하지만 **보정 후에는 일부만 유의**하다고 판단됨
